```python
import pandas as pd
import geopandas as gpd
import psycopg2
from shapely import wkb
from shapely.geometry import mapping as to_geojson
import folium

pd.set_option('display.max_columns', None)
```

```python
def show_poly(geom):
    m = folium.Map(tiles='Stamen Toner')
    folium.GeoJson(
        to_geojson(geom),
        name='Geom',
        style_function=lambda x: {'fillColor': '#0000ff', 'weight': 2, 'color': '#0000ff'}
    ).add_to(m)
    
    m.fit_bounds([(geom.bounds[1], geom.bounds[0]), (geom.bounds[3], geom.bounds[2])])
        
    display(m)
```

```python
df = pd.read_csv("./department_sample_guids.csv").dropna()
conn = psycopg2.connect('service=firecares')
```

```python
sql = """
update firestation_firedepartment
set geom = govt_unit.geom, population = govt_unit.population
from (select geom, population from usgs_governmentunits where id = %(unit_id)s and source = %(source)s) as govt_unit
where id = %(fd_id)s;
"""

with conn.cursor() as c:
    for r in df.to_dict(orient='records'):
        print r['source']
        c.execute(sql, {'fd_id': r['id'], 'unit_id': r['geom_id'], 'source': r['source']})
conn.commit()
```

```python
show_poly(gpd.read_postgis("select * from firestation_firedepartment where id = 73790", conn)['geom'][0])
```
